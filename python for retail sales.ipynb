```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_excel("Pivot_data.xlsx")
df.head()
```

```python
df.columns = ["CustomerID", "TotalSpend", "TotalQuantity", "PurchaseFrequency"]
```

```python
X = df[["TotalSpend", "TotalQuantity", "PurchaseFrequency"]]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
cluster_profile = df.groupby("Cluster")[["TotalSpend", "TotalQuantity", "PurchaseFrequency"]].mean()
print("Cluster profiles:\n")
print(cluster_profile)
```

```python
df.to_csv("customer_segments.csv", index=False)
print("\nSaved customer_segments.csv with cluster labels.")
```
